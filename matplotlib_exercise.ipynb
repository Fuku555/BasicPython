## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

```python
# データセットの読み込み
df_hist = pd.read_csv('data/hist_data.csv')
```

![問1](./imgs/matplotlib/plot_1.png)

```python

# aデータとbデータの度数分布を30分割で表示
plt.hist(df_hist['aデータ'],  bins=30,alpha=0.9, label='aデータ',color='#9FEFE7')
plt.hist(df_hist['bデータ'],  bins=30,alpha=0.7, label='bデータ',color='#FFD17F')
plt.legend()
plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.title('aデータとbデータに関する温度の度数分布')
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

```python
df_scatter = pd.read_csv('data/scatter_data.csv')
```

![問2](./imgs/matplotlib/plot_2.png)

```python
# クラスごとにデータを分割
class_0_data = df_scatter[df_scatter['class'] == 0]
class_1_data = df_scatter[df_scatter['class'] == 1]
class_2_data = df_scatter[df_scatter['class'] == 2]
# 各クラスのデータをプロット
plt.scatter(class_0_data['length (cm)'], class_0_data['width (cm)'], label='Class 0', c='red')
plt.scatter(class_1_data['length (cm)'], class_1_data['width (cm)'], label='Class 1', c='blue')
plt.scatter(class_2_data['length (cm)'], class_2_data['width (cm)'], label='Class 2', c='green')
# プロットのラベル設定
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.legend()
# タイトルの設定
plt.title('クラス0~2のlengthとwidthに関する散歩図')
# グリッド線表示・プロット表示
plt.grid(True)
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

```python
df_line = pd.read_csv('data/line_data.csv')
df_line
```

```python
x=df_line['期間']
y=df_line['件数']
plt.grid(True)
plt.plot(x, y)
plt.xlabel('期間')
plt.ylabel('件数')
plt.title('期間ごとの件数の推移')
plt.xticks(rotation=30)
plt.scatter(x, y, color='#247AB6', s=25)

```

```python
plt.bar(df_line['期間'],df_line['費用'])
plt.xlabel('期間')
plt.ylabel('費用(円)')
plt.title('期間ごとの費用の推移')
plt.xticks(rotation=30)
plt.grid()
```
